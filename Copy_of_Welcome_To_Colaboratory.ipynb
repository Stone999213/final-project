```
class BankAccount(object):
  def __init__(self,owner,balance=0.0,transactions=[]):
    self.owner=owner
    self.balance=balance
    self.transactions = transactions
  
  def __str__(self):
    string="Owner:  "+ self.owner + "\n"
    string+="Balance: $"+ str(self.balance)
    return string
  
  def deposit(self,amount):
    self.balance+=amount
    self.transactions.append(["Deposit",amount])  
  
  def withdrawel(self,amount):
    self.balance-=amount
    if self.balance <0:
      self.balance+=amount
    else:
      self.transactions.append(["Withdrawel",amount])   
  def get_transaction_count(self):
    return len(self.transactions)   

bank_account =BankAccount("Robert")

print(bank_account)
bank_account.withdrawel(25)
print(bank_account)
```

```
from collections import Counter
import numpy as np
math_list = [5,10,10,20,30,25]
clmath_list=[]
for i in math_list:
  clmath_list.append(i)
mean = np.mean(clmath_list) 
median = np.median(clmath_list)
# process for obtaining the mode of the clmath list
def most_frequent(clmath_list):
  occurence_count=Counter(clmath_list)
  return occurence_count.most_common(1)[0][0]
mode = (most_frequent(clmath_list))

print("the mean for the numberical list is ",int(mean))
print("the median for the numerical list is ",median)
print("the mode for the numerical list is ",mode)
```

## Data science

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```
